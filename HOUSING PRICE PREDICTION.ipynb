```python
# Impute the null values in neumerical features of train dataset.
def randomsampleimputation(train_num, variable):
    train_num[variable]=train_num[variable]
    random_sample=train_num[variable].dropna().sample(train_num[variable].isnull().sum(),random_state=0)
    random_sample.index=train_num[train_num[variable].isnull()].index
    train_num.loc[train_num[variable].isnull(),variable]=random_sample
```

```python
randomsampleimputation(train_num, "LotFrontage")
randomsampleimputation(train_num, "MasVnrArea")
randomsampleimputation(train_num, "GarageYrBlt")
```

# Evaluating the Models .

```python
max_r_score=0
r_state=0
for i in range(1,200):
    x_train_pca,x_test_pca,y_train,y_test=train_test_split(xpca,y,test_size=0.25,random_state=r_state)
    reg=LinearRegression()
    reg.fit(x_train_pca,y_train)
    y_pred=reg.predict(x_test_pca)
    r2_scr=r2_score(y_test,y_pred)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        r_state=1
print('Max r2 score :-', max_r_score, "on random state :-",r_state)
```
